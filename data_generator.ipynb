```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
t = np.random.uniform(0.9, 1.1, 500)
```

```python
plt.plot(t)
```

```python
s = np.sin(np.linspace(0, 2*np.pi*6, 500)) * 10 + 20
```

```python
plt.plot(s)
```

```python
r = np.random.uniform(-3, 3, 500)
```

```python
plt.plot(r)
```

```python
idx = np.random.randint(0, 499, 300)
r[idx] = 0
```

```python
y = t + s + r
```

```python
plt.plot(y)
```

```python
y[150] = -0.2
y[420] = 6
y[240:260] = np.random.uniform(19, 20, 20)
```

```python
plt.plot(y)
```

```python
df = pd.DataFrame(y, columns=["data"])
df.to_csv("data/serie_de_mentira.csv", index=None)
```
